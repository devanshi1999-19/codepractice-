```python
# dict comprehension


month=["Jan","Feb","Mar","Apr","May","June","July","Aug","Sep","Oct","Nov","Dec"]

d ={month[i]:i+1 for i in range(len(month))}
d
```

```python
#accessing all key and value pairs
for item in d.items():
  print(item[0],item[1])
```

```python
d.values
```

```python
#accessing all key and value pairs
for item in d.values():
  print(item)

#accessing all key and value pairs
for item in d.keys():
  print(item)
```

```python
import csv
from geopy.distance import distance
stops=list(csv.DictReader(open('stops.txt')))
```

```python
stop_name_list=[row['stop_name'] for row in stops]
len(stop_name_list)
stop_name_list
```
